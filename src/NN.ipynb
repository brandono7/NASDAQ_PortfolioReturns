# Neural Networks (2 Hidden Layers)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
import keras_tuner as kt

# ----------------------------
# 1. Data Loading and Preprocessing
# ----------------------------
file_path_equity = "../data/cleaned_df_with_tickers.csv"
file_path_risk_free = "risk_free_rate.csv"

df_equity = pd.read_csv(file_path_equity)
df_equity['Date'] = pd.to_datetime(df_equity['Date'])

df_risk_free = pd.read_csv(file_path_risk_free)
df_risk_free['Date'] = pd.to_datetime(df_risk_free['Date'])

# Merge equity data with risk-free rate data on Date
df = pd.merge(df_equity, df_risk_free, on='Date', how='left')

# Define features and target variable.
features = df.columns.tolist()
for col in ['Ticker', 'Date', 'risk_premium', 'Risk_Free_Rate']:
    if col in features:
        features.remove(col)
target = 'risk_premium'

# Remove rows with infinite values and drop NaNs
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

# ----------------------------
# 2. Rolling Window Forecasting with Monthly Rebalancing
# ----------------------------
def rolling_window_forecast_nn_portfolio(df):
    years = sorted(df['Date'].dt.year.unique())
    y_actual_all_nn = []
    y_pred_all_nn = []
    portfolio_returns_nn = []
    scaler = StandardScaler()

    for test_year in years:
        # Use test years from 2012 onward
        if test_year < 2012:
            continue

        train_start_date = pd.Timestamp(f"{test_year - 12}-01-01")
        train_end_date = pd.Timestamp(f"{test_year - 2}-12-31")
        val_start_date = pd.Timestamp(f"{test_year - 1}-01-01")
        val_end_date = pd.Timestamp(f"{test_year - 1}-12-31")
        test_start_date = pd.Timestamp(f"{test_year}-01-01")
        test_end_date = pd.Timestamp(f"{test_year}-12-31")

        train_data = df[(df['Date'] >= train_start_date) & (df['Date'] <= train_end_date)]
        val_data = df[(df['Date'] >= val_start_date) & (df['Date'] <= val_end_date)]
        test_data_year = df[(df['Date'] >= test_start_date) & (df['Date'] <= test_end_date)]

        if train_data.empty or val_data.empty or test_data_year.empty:
            continue

        # Scale features using training sample only
        X_train = scaler.fit_transform(train_data[features])
        y_train = train_data[target].values

        X_val = scaler.transform(val_data[features])
        y_val = val_data[target].values

        X_test_year = scaler.transform(test_data_year[features])
        y_test_year = test_data_year[target].values

        # ----------------------------
        # Keras Tuner: Define model-building function with Dropout Regularization
        # ----------------------------
        def build_model(hp):
            model = Sequential()
            model.add(Dense(units=hp.Int('units1', min_value=32, max_value=128, step=32),
                            activation='relu', input_shape=(X_train.shape[1],)))
            model.add(Dropout(rate=hp.Float('dropout_rate1', min_value=0.1, max_value=0.5, step=0.1)))
            model.add(Dense(units=hp.Int('units2', min_value=16, max_value=64, step=16), activation='relu'))
            model.add(Dropout(rate=hp.Float('dropout_rate2', min_value=0.1, max_value=0.5, step=0.1)))
            model.add(Dense(1))  # Output layer for regression
            lr = hp.Float('lr', min_value=1e-4, max_value=1e-2, sampling='log')
            model.compile(optimizer=Adam(learning_rate=lr), loss='mse')
            return model

        tuner = kt.RandomSearch(
            build_model,
            objective='val_loss',
            max_trials=5,
            directory='kt_dir',
            project_name=f'nn_{test_year}',
            overwrite=True
        )
        
        tuner.search(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val), verbose=0)
        
        best_model = tuner.get_best_models(num_models=1)[0]

        X_train_val = np.concatenate([X_train, X_val], axis=0)
        y_train_val = np.concatenate([y_train, y_val], axis=0)
        
        best_model.fit(X_train_val, y_train_val, epochs=50, batch_size=32, verbose=0)

        predictions_year = best_model.predict(X_test_year).flatten()
        
        y_actual_all_nn.extend(y_test_year)
        y_pred_all_nn.extend(predictions_year)

        # Preserve the Ticker column and add daily predicted risk premium
        test_data_year.loc[:, 'predicted_risk_premium'] = predictions_year

        # ----------------------------
        # Monthly Rebalancing using Daily Predictions
        # ----------------------------
        monthly_returns_nn = []
        
        for month in range(1, 13):
            test_data_month = test_data_year[test_data_year['Date'].dt.month == month]
            if test_data_month.empty:
                continue

            # Group stocks into deciles based on predicted risk premiums.
            test_data_month.loc[:, 'decile'] = pd.qcut(
                test_data_month['predicted_risk_premium'].rank(method='first'),  
                q=10,
                labels=False,
                duplicates="drop"
            )

            top_decile_returns = test_data_month[test_data_month['decile'] == 9]['risk_premium'].mean()
            bottom_decile_returns = test_data_month[test_data_month['decile'] == 0]['risk_premium'].mean()
            
            long_short_return_month = top_decile_returns - bottom_decile_returns
            
            avg_risk_free_rate_month = test_data_month['Risk_Free_Rate'].mean()
            
            monthly_returns_nn.append({
                'Month': month,
                'Long-Short Return': long_short_return_month,
                'Risk-Free Rate': avg_risk_free_rate_month,
                'Excess Return': long_short_return_month  
            })

        if monthly_returns_nn:
            monthly_df_nn = pd.DataFrame(monthly_returns_nn)
            
            yearly_cumulative_return_nn = (1 + monthly_df_nn['Long-Short Return']).prod() - 1
            
            sharpe_ratio_nn = (
                monthly_df_nn['Excess Return'].mean() / monthly_df_nn['Excess Return'].std()
                if monthly_df_nn['Excess Return'].std() != 0 else 0
            )
            
            portfolio_returns_nn.append({
                'Year': test_year,
                'Yearly Cumulative Return': yearly_cumulative_return_nn,
                'Sharpe Ratio': sharpe_ratio_nn,
            })

    return {
        "y_actual": y_actual_all_nn,
        "y_pred": y_pred_all_nn,
        "portfolio_returns": portfolio_returns_nn,
    }

# ----------------------------
# Main Execution Code
# ----------------------------
results_nn = rolling_window_forecast_nn_portfolio(df)

portfolio_df_nn = pd.DataFrame(results_nn['portfolio_returns'])
portfolio_df_nn.to_csv('NN_portfolio_performance.csv', index=False)

dm_test_df_nn = pd.DataFrame({
    'y_actual': results_nn['y_actual'],
    'y_pred': results_nn['y_pred']
})
dm_test_df_nn.to_csv('dm_test_results_NN.csv', index=False)

print("[Performance Metrics]")
print(f"Sharpe Ratio (Mean): {portfolio_df_nn['Sharpe Ratio'].mean():.2f}")
print(f"Final Cumulative Return: {portfolio_df_nn['Yearly Cumulative Return'].iloc[-1]}")
print("Overall R^2:", r2_score(results_nn['y_actual'], results_nn['y_pred']))

plt.figure(figsize=(10, 5))
plt.plot(portfolio_df_nn['Year'], portfolio_df_nn['Yearly Cumulative Return'], marker='o', label='Cumulative Return')
plt.xlabel('Year')
plt.ylabel('Cumulative Return')
plt.title('Neural Network Portfolio Cumulative Returns Over Time')
plt.legend()
plt.grid()
plt.show()
```
